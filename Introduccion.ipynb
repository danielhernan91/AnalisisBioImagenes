# GENERALIDADES

* Las imágenes están compuestas por píxeles. La palabra 'píxel' deriva de la expresión inglesa picture element (elemento de imagen)
* Las imágenes que parecen iguales pueden contener diferentes valores de píxeles.
Las imágenes que parecen diferentes aún pueden contener los mismos valores de píxeles. 8bits,16bits,32bits
* Tablas de búsqueda (LUTs). Las tablas de búsqueda a veces se denominan mapas de colores. El software que muestra la imagen utiliza una tabla de búsqueda (LUT) que asigna cada valor de píxel a un color.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 
import matplotlib.patches as patches

# Cargar una imagen y seleccionar una región detallada
im = mpimg.imread('imagen.tif') # Use mpimg.imread()
x, y, width, height = (205, 100, 120, 120)
im_detail = im[y:y+height, x:x+width]

# Mostrar gráficos
fig = plt.figure(figsize=(8, 4)) 

# Crear un rectángulo para resaltar la región seleccionada
rect = patches.Rectangle((x, y), width, height, linewidth=1, edgecolor='r', facecolor='none')

# Mostrar la imagen original
plt.subplot(131)
plt.imshow(im, vmin=0, vmax=255) 
plt.title('(A) Imagen original')
plt.gca().add_patch(rect) # Añadir el rectángulo a la imagen original

# Mostrar la vista ampliada de la región seleccionada
plt.subplot(132)
plt.imshow(im_detail, vmin=0, vmax=255)
plt.title('(B) Vista ampliada de (A)')

# Mostrar los valores de píxeles de la región ampliada
plt.subplot(133)
plt.imshow(im_detail, vmin=0, vmax=255, alpha=1)
plt.title('(C) Valores de píxeles de (B)')

# Anotar cada píxel con su valor numérico
for (i, j), z in np.ndenumerate(im_detail):
    plt.text(j, i, str(z), ha='center', va='center', fontdict={'size': 5})

plt.tight_layout() 
plt.show() 
```
